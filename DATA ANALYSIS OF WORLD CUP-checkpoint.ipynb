```python
argentina_data = ft[ft['Team'] == 'Argentina']
```

```python
print(argentina_data.describe())
```

# Argentinas record about goals

```python
sns.lineplot(x='year', y='Goals For', data=argentina_data, label='Goals For')
sns.lineplot(x='year', y='Goals Against', data=argentina_data, label='Goals Against')
plt.title('Argentina Goals Over Years')
plt.legend()
plt.show()
```

# Argentinas Win percentage

```python
argentina_data['Win Percentage'] = argentina_data['Win'] / argentina_data['Games Played'] * 100
print(argentina_data[['year', 'Win Percentage']])
```

# Argentina's average goals for vs goal against

```python
argentina_data['Avg Goals For'] = argentina_data['Goals For'] / argentina_data['Games Played']
argentina_data['Avg Goals Against'] = argentina_data['Goals Against'] / argentina_data['Games Played']
print(argentina_data[['year', 'Avg Goals For', 'Avg Goals Against']])
```

```python
sns.lineplot(x='year', y='Points', data=argentina_data)
plt.title('Argentina Performance Over Years')
plt.show()
```
